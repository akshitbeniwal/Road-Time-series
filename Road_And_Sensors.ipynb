```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn import preprocessing
```

```python
df = pd.read_csv('dataset.csv')
```

```python
df.head()
```

```python
df = df.drop(['Unnamed: 0'], axis = 1)
```

```python
df_train = df[df.year != 10]
```

```python
df_train.tail()
```

```python
df_test = df[df.year == 10]
```

```python
df_test.head()
```

```python
scaler = preprocessing.StandardScaler()
df_train = df_train.drop(['year'], axis = 1)
df_test = df_test.drop(['year'], axis = 1)
df_train = scaler.fit_transform(df_train)
df_test = scaler.transform(df_test)
```

```python
df_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model_new = Sequential()
x1 = df_train[:, :8]
y1= df_train[:, 8]
model_new.add(Dense(10, activation = "relu", input_shape = (8,)))
model_new.add(Dense(10, activation = "relu"))
model_new.add(Dense(5, activation = "relu"))
model_new.compile(optimizer = "adam", loss = "mean_squared_error", metrics =[tf.keras.metrics.RootMeanSquaredError()])
model_new.fit(x1,y1, epochs = 30, batch_size = 10)
```

```python
df
```

```python
x2 = df_test[:,:8]
y2 = df_test[:, 8]


model_new.evaluate(x2, y2)
```
